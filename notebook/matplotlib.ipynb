```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## hist

```python
x1 = np.random.normal(2.0, 2.0, 10000)
x2 = np.random.normal(0.0, 1.0, 10000)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
x_min, x_max = -5, 11
ax.hist(x1, bins=50, range=(x_min, x_max), alpha=0.5, label="x1", density=True, color="C0")
ax.hist(x2, bins=50, range=(x_min, x_max), alpha=0.9, label="x2", density=True, color="C1", 
        histtype="step")
ax.legend()
ax.grid()
plt.show()
```

## bar

```python
d1 = [1.0, 2.0, 3.0, 4.0]
d2 = [5.0, 4.5, 3.5, 2.5]
ticks = ['AAA', 'BBB', 'CCC', 'DDD']
n_kinds = len(ticks)
x = np.arange(n_kinds)
w = 0.2

fig, ax = plt.subplots(1, 1)
ax.bar(x,   d1, width=w, log=False, alpha=0.8, color="C0", label="train")
ax.bar(x+w, d2, width=w, log=False, alpha=0.8, color="C3", label="test")
ax.set_title(f"bar")
ax.axes.set_xticks(x+w/2)
ax.axes.set_xticklabels(ticks)
ax.tick_params(axis='x', rotation=45)
ax.legend(loc=0)
ax.grid(axis='y')
plt.show()
```

## scatter

```python
x = np.random.normal(2.5, 1.0, 1000)
y = np.random.normal(0.5, 1.0, 1000)
fig, ax = plt.subplots(1, 1)
ax.scatter(x, y, s=2)
ax.set_aspect('equal', adjustable='box')
ax.axhline(0.5, color="C1")
ax.axvline(2.5, color="C3")
ax.grid()
plt.show()
```

## Time Series

```python
df = pd.DataFrame(np.random.randn(60, 1), 
                  columns=['A'], 
                  index=pd.date_range('20130101', periods=60, freq='M'))

fig, ax = plt.subplots(1, 1)
df.plot(ax=ax)
ax.axvspan("2015-01-01", "2016-01-01", color="r", alpha=0.2) # 
ax.grid()
plt.show()
```
